```python
# Import the driver and numpy
from src.correlation_driver import Correlation
import numpy as np
```

```python
# Set up files, tracer type, and params
lymanalphaCorr = Correlation(data_file = 'dr16_NGC_eboss_lyalpha.fits',
                             data_type = 'lyman_alpha_1D',
                             cfg_file = 'ex_par.txt',
                             verbose = True)
```

```python
# Partition the catalog
lymanalphaCorr.divide(save_plan = True,
                      plot_results = True,
                      save_plot = False,
                      theta_p = 13.)
```

```python
# Convolve using 2pcf mu wedge mode with desired mu binning
lymanalpha_mu_bin_edges = np.asarray([0.0,0.5,0.8,0.95,1.0])
lymanalphaCorr.convolve_2pcf_mu_wedges(mu_edges = lymanalpha_mu_bin_edges)
```

```python
# Sum over the resulting grids
lymanalphaCorr.wrap_2pcf_mu_wedges(mu_edges = lymanalpha_mu_bin_edges, 
                                   plot_correlation = False,
                                   save_correlation_plot = False)
```
