```python
# Import the driver and numpy
from src.correlation_driver import Correlation
import numpy as np
```

```python
# Set up files, tracer type, and params
galaxyCorr = Correlation(data_file = 'dr12_NGC_cmass_galaxies.fits',
                         data_type = 'galaxy',
                         cfg_file = 'ex_par.txt',
                         randoms_file = 'dr12_NGC_cmass_randoms.fits',
                         verbose = True)
```

```python
# Partition the catalog
galaxyCorr.divide(save_plan = True,
                  plot_results = True,
                  save_plot = True)
```

```python
# Convolve using 2pcf mu wedge mode with desired mu binning
galaxy_mu_bin_edges = np.linspace(0.,1.,3)
galaxyCorr.convolve_2pcf_mu_wedges(mu_edges = galaxy_mu_bin_edges,
                                   store_rand = True)
```

```python
# Sum over the resulting grids
galaxyCorr.wrap_2pcf_mu_wedges(mu_edges = galaxy_mu_bin_edges, 
                               plot_correlation = False,
                               save_correlation_plot = False)
```
